# Normalize

```python
from sklearn.preprocessing import StandardScaler

colunas_selecionadas = [
    'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES',
    'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE',
    'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY',
    'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY',
    'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS',
    'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE'
]

df_selecionado = dados[colunas_selecionadas]
scaler = StandardScaler()
df_normalizado = df_selecionado.copy()
df_normalizado[colunas_selecionadas] = scaler.fit_transform(df_normalizado[colunas_selecionadas])
```

### Matrix de correlação normalizada

```python
corrmat = df_normalizado.corr()
corrmat
```

# KMeans

```python
from sklearn.cluster import KMeans

inertias = []
for i in range(1,10):
  # kmeans++ ira selecionar os clusters baseado em dristribuição de probabilidade empírica
  # dos pontos, ela acelera a convergencia do método.
  km = KMeans(n_clusters=i, init = 'k-means++', random_state=321, n_init='auto')
  km.fit(df_normalizado)
  inertias.append(km.inertia_)

plt.figure(figsize=(12,6))
plt.plot([x for x in range(1,10)], inertias, marker='o')
plt.xlabel('K')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

inertias = []

for i in range(2, 11):  # Silhouette score não faz sentido para 1 cluster
    km = KMeans(n_clusters=i, init='k-means++', random_state=321, n_init='auto')
    km.fit(df_normalizado)
    silhouette_avg = silhouette_score(df_normalizado, km.labels_)
    print(f'Silhouette Score for {i} clusters = {silhouette_avg:.6f}')
    inertias.append(km.inertia_)
```

```python
from sklearn.cluster import KMeans

# Aplicar o KMeans com 3 clusters
km_3_clusters = KMeans(n_clusters=3, init='k-means++', random_state=321, n_init='auto')
km_3_clusters.fit(df_normalizado)

# Adicionar os rótulos dos clusters ao DataFrame original
dados['ClusterKMeans'] = km_3_clusters.labels_

# Mostrar os primeiros registros do DataFrame com os clusters
dados.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_normalizado)

dados['PCA1'] = df_pca[:, 0]
dados['PCA2'] = df_pca[:, 1]

# Plotar os clusters visualmente
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='ClusterKMeans', data=dados, palette='viridis', s=100, alpha=0.7)
plt.title('Visualização dos Clusters com PCA')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Aplicar PCA para reduzir para 3 dimensões
pca_3d = PCA(n_components=3)
df_pca_3d = pca_3d.fit_transform(df_normalizado)

dados['PCA1'] = df_pca_3d[:, 0]
dados['PCA2'] = df_pca_3d[:, 1]
dados['PCA3'] = df_pca_3d[:, 2]


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


sc = ax.scatter(dados['PCA1'], dados['PCA2'], dados['PCA3'], c=dados['ClusterKMeans'], cmap='viridis', s=50, alpha=0.7)


ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.title('Visualização dos Clusters em 3D com PCA')
plt.colorbar(sc)
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1.8, min_samples=10)
dblabels = dbscan.fit_predict(df_normalizado)

dados['ClusterDB'] = dblabels
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_normalizado)

dados['PCA1'] = df_pca[:, 0]
dados['PCA2'] = df_pca[:, 1]

# Plotar os clusters visualmente
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='ClusterDB', data=dados, palette='viridis', s=100, alpha=0.7)
plt.title('Visualização dos Clusters com PCA')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


sc = ax.scatter(dados['PCA1'], dados['PCA2'], dados['PCA3'], c=dados['ClusterDB'], cmap='viridis', s=50, alpha=0.7)


ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.title('Visualização dos Clusters em 3D com PCA')
plt.colorbar(sc)
plt.show()
```

```python
silhouette_avg = silhouette_score(df_normalizado, dados['ClusterKMeans'])
print(f'Silhouette Score para KMeans: {float(silhouette_avg):.4f}')
```

```python
silhouette_avg = silhouette_score(df_normalizado, dados['ClusterDB'])
print(f'Silhouette Score para DBSCAN: {float(silhouette_avg):.4f}')
```

# HDBSCAN

```python
from sklearn.cluster import HDBSCAN

for i in range(2, 8):
    hdb = HDBSCAN(min_cluster_size=i)
    hdb.fit(df_normalizado)

    dados['ClusterHDB'] = hdb.labels_

    silhouette_avg = silhouette_score(df_normalizado, dados['ClusterHDB'])
    print(f'Silhouette Score para {i} Cluster-Size: {float(silhouette_avg):.4f}')
```

```python
hdb = HDBSCAN(min_cluster_size=4)
hdb.fit(df_normalizado)

dados['ClusterHDB'] = hdb.labels_
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='ClusterHDB', data=dados, palette='viridis', s=100, alpha=0.7)
plt.title('Visualização dos Clusters com PCA')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


sc = ax.scatter(dados['PCA1'], dados['PCA2'], dados['PCA3'], c=dados['ClusterHDB'], cmap='viridis', s=50, alpha=0.7)


ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.title('Visualização dos Clusters em 3D com PCA')
plt.colorbar(sc)
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score
```

```python
# Davies-Bouldin Score (menores valores, melhor cluster)
dbi = davies_bouldin_score(df_normalizado, dados['ClusterHDB'])
print(f"Davies-Bouldin Index: {dbi:.2f}")

# Calinski-Harabasz Index (maiores valores, melhor cluster)
ch_index = calinski_harabasz_score(df_normalizado, dados['ClusterHDB'])
print(f"Calinski-Harabasz Index: {ch_index:.2f}")
```
