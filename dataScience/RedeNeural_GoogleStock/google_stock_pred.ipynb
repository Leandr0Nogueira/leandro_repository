```python
selected_columns = ['close', 'high', 'low', 'open', 'volume', 'adjClose', 'adjHigh',
       'adjLow', 'adjOpen', 'adjVolume']
```

```python
from sklearn.preprocessing import StandardScaler

df_selecionado = dados[selected_columns]
scaler = StandardScaler()
df_normalizado = df_selecionado.copy()
df_normalizado[selected_columns] = scaler.fit_transform(df_normalizado[selected_columns])

df_normalizado['date'] = pd.to_datetime(dados['date'])
```

```python
df_normalizado.head()
```

```python
X = df_normalizado.drop(columns=['open'])
y = df_normalizado.open
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import cross_val_score

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=21, test_size=0.2)

print('Treinaremos com %d elementos e testaremos com %d elementos' % (len(X_train), len(X_test)))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Definindo o modelo LSTM
model = Sequential()

# Primeira camada LSTM
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

# Segunda camada LSTM
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

# Camada densa
model.add(Dense(units=25))

# Camada de sada
model.add(Dense(units=1))

# Compilando o modelo
model.compile(optimizer='adam', loss='mean_squared_error')
```
