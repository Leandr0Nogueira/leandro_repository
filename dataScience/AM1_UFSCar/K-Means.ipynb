```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Gerar um conjunto de dados bidimensional
X, y = make_blobs(n_samples=50, centers=10, n_features=2, random_state=36)

# Função para plotar os dados e os centroids
def plot_kmeans(X, centroids, labels, iteration):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x')
    plt.title(f'Iteração {iteration}')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()

# Inicializar os centroids
k = 3
centroids = X[np.random.choice(X.shape[0], k, replace=False)]

# Iterar o K-Means
max_iterations = 10
for iteration in range(max_iterations):
    # Atribuir cada ponto ao centroide mais próximo
    distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
    labels = np.argmin(distances, axis=1)
    
    # Plotar a iteração atual
    plot_kmeans(X, centroids, labels, iteration)
    
    # Recalcular os centroids
    new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(k)])
    
    # Verificar se os centroids mudaram
    if np.all(centroids == new_centroids):
        break
    
    centroids = new_centroids
```
